```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
for i in range(1, 10):
    
    df = pd.read_csv('data/exp_1_1_data_{}.csv'.format(i))[['y']+[str(i) for i in range(0, 70)]]
    view = df.groupby(['y']).mean().T
    df = pd.read_csv('data/exp_1_1_data_{}.csv'.format(i))[['y']+['fake_'+str(i) for i in range(0, 70)]]
    view_f = df.groupby(['y']).mean().T
    plt.figure(figsize=(15, 7))
    plt.title("probabilités données aux classes vues pendant l'enrtainement, aux classes jamais vues, et aux images du générateur")
    plt.plot(view[[i for i in range(i)]].mean(1), label='Seen Training Classes proba: {}'.format(str([j for j in range(i)])))
    plt.plot(view[[i for i in range(i, 10)]].mean(1), label='Unseen Training Classes proba: {}'.format(str([j for j in range(i, 10)])))
    plt.plot(list(view_f[0]), label='Generated samples probability')

    plt.legend()
    plt.show()
    plt.figure(figsize=(15, 7))
    plt.title("Différence de probabilités donnée par le discriminateur aux images réelles \n entre classes vues pendant l'entrainement et classes jamais vues")
    plt.plot(view[[i for i in range(5)]].mean(1)-view[[i for i in range(5, 10)]].median(1))
    plt.show()
    print('\n\n---------------------------------------------------------------------------------------------------------\n\n')
    print('---------------------------------------------------------------------------------------------------------\n\n')
```
